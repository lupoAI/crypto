```python
import pandas as pd
import json
```

```python
with open('../data/exchange/arbitrage_paths.json', 'r') as f:
    arbitrage_paths = json.load( f)
```

```python
tickers_to_trade = {}
for coin in list(arbitrage_paths.keys()):
    c = arbitrage_paths[coin]
    if len(c)==0:
        continue
    tickers_to_trade[coin] = []
    for i, path in enumerate(c):
        c[i] = [coin] + path
        temp = []
        for q, b in zip(c[i][:-1], c[i][1:]):
            temp += [b+q]
        tickers_to_trade[coin] += [temp]
```

```python
import os
base_path = r"G:\crypto\data\symbols"
exchange_info = pd.read_csv("../data/exchange/exchangeInfo.csv", index_col=[0])
exchange_info.head()
```

```python
have_data = {}
for ticker in exchange_info['symbol'].values:
    have_data[ticker] = os.path.exists(os.path.join(base_path, f"{ticker}_1m.csv"))
```

```python
def all_have_data(l):
    for ticker in l:
        has_data = have_data[ticker]
        if not has_data:
            return False
    return True
    
```

```python
for coin in list(tickers_to_trade.keys()):
    tickers_to_trade[coin] = list(filter(all_have_data, tickers_to_trade[coin]))
```
